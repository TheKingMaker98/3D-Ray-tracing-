![monteCarlo](https://i.imgur.com/cORgbDD.png)
Remarque : La proportion de points Verts divisée par 1000 donne une approximation de la probabilité pour un point tiré au hasard dans le carré de côté 2 de se trouver dans le cercle centré en 0 et de rayon 1.
Cette probabilité est égale à $$P=\frac{Aire_{disque}}{Aire_{carre}} = \frac{\pi \times 1^2}{4} = \frac{\pi}{4}$$
Vérifier que la proportion de points Verts multipliée par $4$ donne bien une valeur proche de $\pi$.

Cette méthode pour déterminer de façon probabiliste la valeur de $\pi$ est la méthode de Monte-Carlo.
Elle n'est précise que pour un très grand nombre de points (et encore...).

## 3) Tracer la courbe représentative d'une fonction

Pour tracer une courbe, une calculatrice ou un ordinateur ne calcule pas les coordonnées de tous les points de la courbe (il y en a une infinité). Ils se contentent de calculer régulièrement des valeurs de la fonction puis les relie des segments qui ne se voient presque pas à l'oeil nu (lorsqu'il y a beaucoup de valeurs). C'est ce que nous allons faire.


Le tracé d'une fonction s'obtient à partir de son tableau de valeurs qui regroupe les coordonnées d'un certain nombre de points de la courbe à intervalles réguliers. On appelle "pas" l'écart régulier entre deux valeurs successives de x.

Dans notre cas, plutôt que de fixer un pas on va fixer ***le nombre de valeurs*** que l'on va calculer pour faire notre tracé.

Voici comment nous allons procéder : 

1. A l'aide de la fonction ```linspace```, on commence par construire le vecteur d'abscisses contenant l'ensemble des valeurs dont on va calculer l'image .

2. On construit les images des valeurs précédemment calculées en appliquant la fonction sur le vecteur. 

3. On passe sur la partie technique, préparer le graphique, placer les axes, les graduations, le titre...

Etudier l'exemple suivant, il contient tout ce qu'il y a à savoir pour obtenir un beau graphique...

````python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Bornes
xMin = -4
xMax = 4

# Nombre de points
N = 100

# 1. il nous faut un vecteur de N valeurs de x, équi-réparties entre xMin et xMax
xs = np.linspace(xMin,xMax,N) 
        
# la fonction que l'on souhaite tracer
def f(x):
    # la fonction que l'on veut tracer
    return x * x 
    
# 2. on construit le tableau des f(x) pour chaque élement du tableau d'abscisses
ys=f(xs)
    
# 3. Ajoute la courbe
plt.plot(xs,ys, color = "red", linewidth=3, label="$x^2$")

# paramètre les axes
plt.xlim([-5, 5])
plt.ylim([-1, 10])

# ax reçoit "get the current axes : gca"
ax = plt.gca()
# et on les modifie
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.5))
plt.axhline(color = 'blue', linewidth=1) # axe des abscisse
plt.axvline(color='blue', linewidth=1) # axe des ordonnées

# Ajout les noms des axes et le titre
plt.xlabel("x")
plt.ylabel("y")
plt.title("La fonction carré")
# ajout de la légende (en haut à droite)
ax.legend(loc='upper right')

plt.show() # affiche l'image à l'écran
````

## 3) Tracer un diagramme en bâtons

Tracer un diagramme en baton n'est pas plus difficile que tracer un nuage de points.

Il faut simplement noter qu'un diagramme en bâtons et profondément lié à l'axe des abscisses. 

````python
import matplotlib.pyplot as plt

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

boissons = ['Chocolat', 'Café', 'Thé', "Jus d'orange", 'Vin blanc']
# pour pouvoir maitriser l'emplacement de la barre
ypos = [1,2,3,4,5]
effectif = [23,17,12,31,1]
taille = 0.75
ax.bar(ypos,effectif, taille, color = "blue", align='center')
# on modifie les graduations
plt.xticks(ypos, boissons)
plt.ylabel('Effectif')
plt.title('Boisson au petit déjeuner')

plt.show()
````

## 4) Une animation

Le module animation est disponible à partir de la version 1.1 de matplolib.

Nous allons utiliser la fonction ```FuncAnimation()``` du module animation.

Dans notre script, nous allons définir une fonction ```init()``` et une fonction ```animate()```. La fonction ```init()``` servira à créer l’arrière de l’animation qui sera présent sur chaque image. La fonction animate() met à jour la courbe pour chaque image.

On va travailler exactement comme si on traçait un graphique, mais à chaque appel à ```animate``` on va modifier la liste des points qui la constituent.

Plutôt qu'un long discours...

````python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy # Pour le cosinus !

NB_FRAMES = 100

# mets à jour les y en fonction des xs, mais surtout en fonction de t ! 
def misAjour(xs, t):
    t = t / NB_FRAMES # entre  0 et 1
    ys = []
    for x in xs :
        ys.append(t * numpy.cos(x + 7 * t))
    return ys

# préparation des asbcisses
xs = []
N = 100
for i in range(N):
    xs.append(i * 2*3.15/N)
    
fig, ax = plt.subplots()
#paramètre les axes
plt.xlim([0, 6])
plt.ylim([-1, 1])
# on commence par un tableau vide
l, = ax.plot([],[])

def animate(i):
    l.set_data(xs, misAjour(xs,i))

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=NB_FRAMES)

from IPython.display import HTML
HTML(ani.to_jshtml())
````
